```python
# Step 1: Import all necessary libraries
import numpy as np  # For numerical operations
import pandas as pd # For data manipulation and analysis
import matplotlib.pyplot as plt # For creating visualizations
import seaborn as sns # For nicer visualizations


from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler


%matplotlib inline

print("All libraries imported successfully!")
```

```python
# Step 2: Load the data
# Note: You need to UPLOAD the file to Colab first using the folder icon on the left.
# If the file is in the same folder as your notebook, just use the name.

from google.colab import files
uploaded = files.upload()  # This will prompt you to select a file from your computer. Use this line in Colab.

# Read the uploaded file into a Pandas DataFrame
df = pd.read_csv('Mall_Customers.csv') # Replace with your filename

# Let's see what the data looks like (Display the first 5 rows)
print("Data loaded successfully! Here's a preview:")
df.head()
```

```python
# Step 3: Explore and Prepare Data

# Check for missing values
print("Missing values in each column:")
print(df.isnull().sum())

# Let's choose the features we want to use for clustering
# We'll use Annual Income and Spending Score
X = df[['Annual Income (k$)', 'Spending Score (1-100)']]

# It's often a good idea to scale the data so one feature doesn't dominate
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

print("\nData is prepared and scaled!")
```

```python
# Step 4: Find the optimal k using the Elbow Method

wcss = [] # This list will hold the "within-cluster sum of squares" for different k values

# Let's try k values from 1 to 10
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_) # .inertia_ is the WCSS value

# Now plot the graph
plt.figure(figsize=(10,6))
plt.plot(range(1, 11), wcss, marker='o', linestyle='--')
plt.title('The Elbow Method')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('WCSS')
plt.xticks(range(1,11))
plt.grid(True)
plt.show()
```

```python
# Step 5: Apply K-Means with the chosen k
kmeans = KMeans(n_clusters=5, init='k-means++', random_state=42)

# Fit the model and predict the cluster for each customer
clusters = kmeans.fit_predict(X_scaled)

# Add this cluster info back to our original dataframe
df['Cluster'] = clusters

print("Clustering complete! Cluster labels added to the data.")
df.head()
```

```python
# Step 6: Visualize the clusters

plt.figure(figsize=(12,8))

# Scatter plot: X-axis: Annual Income, Y-axis: Spending Score
# Color the points (c=...) based on their cluster label
scatter = plt.scatter(df['Annual Income (k$)'], df['Spending Score (1-100)'], c=df['Cluster'], cmap='viridis', s=100)

plt.title('Customer Segments')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.colorbar(scatter, label='Cluster Label')
plt.grid(True)
plt.show()

print("Awesome! You've successfully segmented the customers.")
```
